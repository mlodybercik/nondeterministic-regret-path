```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
plt.rcParams["text.usetex"] = True
plt.rcParams["text.latex.preamble"] = "\\usepackage{polski}"
plt.rcParams["figure.figsize"] = (10,3)
plt.rcParams["figure.autolayout"] = False
plt.rcParams["savefig.pad_inches"] = 0.01
plt.rcParams["axes.axisbelow"] = True
plt.rcParams["font.family"] = "sans-serif"
plt.rcParams["font.sans-serif"] = "Helvetica"

node_draw_params = {"node_color": "#FFFFFF"}
```

```python
G = nx.Graph()
G.add_edge(0, 1, label="$\\left(20-50\\right)$")
G.add_edge(0, 2, label="$\\left(12-400\\right)$")
G.add_edge(0, 3, label="$\\left(13-200\\right)$")
G.add_edge(0, 4, label="$\\left(14-20\\right)$")
G.add_edge(0, 5, label="$\\left(19-20\\right)$")

G.add_edge(1, 6, label="$\\left(10-200\\right)$")
G.add_edge(2, 6, label="$\\left(10-2002\\right)$")
G.add_edge(3, 6, label="$\\left(10-23\\right)$")
G.add_edge(4, 6, label="$\\left(10-24\\right)$")
G.add_edge(5, 6, label="$\\left(22-23\\right)$")

pos = {
    0: (-1, 0),
    6: ( 1, 0),
    1: ( 0, 1),
    2: ( 0, 0.5),
    3: ( 0, 0),
    4: ( 0, -0.5),
    5: ( 0, -1),
}

nx.draw(G, pos)
nx.draw_networkx_nodes(G, pos, **node_draw_params)
edge_labels = dict([((n1, n2), d['label']) for n1, n2, d in G.edges(data=True)])
ax = plt.gca()

ax.collections[0].set_edgecolor("#000000")
ax.collections[0].set_linewidth(2)

nx.draw_networkx_edge_labels(G, pos, edge_labels)
nx.draw_networkx_labels(G, pos)
plt.savefig('../sprawko/figures/example.pdf', bbox_inches='tight')
```
